```python
import sklearn #machine learning
import numpy as np # computations
import pandas as pd # data management
```

```python
#1. Load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/china-srun/Machine-Learning-Data-Wrangling/main/Top_650_FIFA.csv")
dataset.head(5)
```

```python
#dimension: rows, colums
dataset.shape
```

```python
#describe the dataset
dataset.describe()
```

```python
#check the variable (numerical / categorical)
numerical_variable = dataset.columns[dataset.dtypes != 'object'] # look for numerical data
dataset[numerical_variable].head(5)
```

```python
#check the variable (numerical / categorical)
categorical_variable = dataset.columns[dataset.dtypes == 'object'] # look for categorical data
dataset[categorical_variable].head(5)
```

```python
#missing values
dataset.isnull().head(5)

#False: no missing values
#True: we have the missing values
```

```python
#find out the length of the dataset
len(dataset)
```

```python
# find out the percentage of missing value in numerical variables
dataset[numerical_variable].isnull().sum() / len(dataset)
```

```python
# find out the percentage of missing value in categorical variables
dataset[categorical_variable].isnull().sum() / len(dataset)
```

```python
#replace
dataset['Position'].replace('Forward', 'forward')
dataset.replace('Midfield', 'middle').head(5)
dataset['age'].replace([34, 29], [38, 25]).head(15)
```

```python
# create the rows and add it with the original dataset
row_insert = pd.Series(['S. China', 'Srun Junior China the Third', 69, 18, '2003-06-19'], index = ['short_name', 'long_name', 'overall', 'age', 'dob'])
dataset.append(row_insert, ignore_index=True)
```

```python
#call the row
#datasetname.iloc[] - we can call the specific row based on the index
dataset.iloc[0]
dataset.iloc[:15]
dataset.iloc[-1]
```

```python
#loc - to call the specific values,
#index
dataset = dataset.set_index(dataset['Position'])
dataset.loc['Forward']
```

```python
#rename
dataset.rename(columns={'dob' : 'date of birth'}).head(5)
```

```python
#find the max value of a column
print("Oldest player's age is ", dataset['age'].max(), "years old.")
```

```python
#delete col
dataset.drop('age', axis=1).head(5)
```

```python
#delete row
dataset.drop("Midfield", axis = 0).head(5)
```

```python
#unique value of a column
dataset['attacking'].unique()
```
